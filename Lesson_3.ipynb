## 1. Постройте нейронную сеть (берем простую линейную сеть, которую разбирали на уроке: меняем число слоев, число нейронов , типы активации, тип оптимизатора) на датасет from sklearn.datasets import load_boston.

2. Измените функцию потерь и метрику для этой задачи. Постройте 10-15 вариантов и сведите результаты их работы в таблицу Опишите, какого результата вы добились от нейросети? Что помогло вам улучшить ее точность?

```python
X, y = load_boston(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=24)
```

```python
X_train = (X_train - np.mean(X_train, axis=0)) / np.std(X_train, axis=0)
X_test = (X_test - np.mean(X_test, axis=0)) / np.std(X_test, axis=0)
```

```python
dict_history_model_boston_house_train = {}
results = {}
```

```python
def model_b(layers_count, neurons_count, opt):
    for i in range (0,layers_count-1):
        neur=neurons_count
        md = Sequential([tf.keras.layers.Dense(neur, activation=tf.nn.relu)])
    md.add(Dense(1, activation='relu'))
    md.compile(optimizer=opt, loss=tf.keras.losses.MeanAbsoluteError(), metrics=['mae'])
    md.fit(X_train, y_train, batch_size=8, validation_split=0.20, epochs=20)
    loss_mae = md.evaluate(X_test, y_test)
    results['MODEL' + '_' + str(layers) + '_' + str(neur)+ '_'+opt] = loss_mae[1]
    print(f"Слоев: {layers_count}, нейронов в слое: {neur}")
    return md
```

```python
for layers in (2, 5):
    for neur in (128, 256, 512):
        for optim in ('adam', 'rmsprop', 'nadam'):
            print(layers, neur, optim)
            model_b(layers, neur, optim)
```

```python
res = pd.DataFrame(results.items(), columns=['Name', 'MAE'])
res
```

```python
imdb = keras.datasets.imdb
```

```python
(train_vec, train_labels), (test_vec, test_labels) = imdb.load_data()
```

```python
type(train_vec)
```

```python
print(train_vec.shape)
print(test_vec.shape)
```

```python
# Приведение векторов к одной длине. Средняя длина 218. Для ровного счета возьмем 256
train_data = keras.preprocessing.sequence.pad_sequences(train_vec, padding='post', maxlen=256)
test_data = keras.preprocessing.sequence.pad_sequences(test_vec, padding='post', maxlen=256)
```

```python
len(train_data[0]), len(train_data[1])
```

```python
print(train_data[0])
```

```python
model = keras.Sequential()
model.add(keras.layers.Embedding(400289, 16))
model.add(keras.layers.GlobalAveragePooling1D())
model.add(keras.layers.Dense(32, activation=tf.nn.relu))
model.add(keras.layers.Dense(16, activation=tf.nn.relu))
model.add(keras.layers.Dense(1, activation=tf.nn.sigmoid))

model.summary()
```

```python
model.compile(optimizer='Adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(train_data, train_labels, batch_size=256, epochs=4)
```

```python
results = model.evaluate(test_data, test_labels)

print(results)
```
